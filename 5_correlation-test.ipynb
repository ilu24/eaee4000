# Data Processing

```python
import pandas as pd
import numpy as np

df = pd.read_csv("FPA-FOD_reduced.csv", low_memory = False)
```

# Correlation Test

```python
numeric_data = df.select_dtypes(include=['float', 'int'])

# Calculate correlation matrix on numeric columns only
correlation_matrix = numeric_data.corr().abs()

print(correlation_matrix)
```

```python
# Print pairs of variables with correlation > 0.7
threshold = 0.7
high_corr = (correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))
             .stack()
             .loc[lambda x: x > threshold])

print(high_corr)
```
